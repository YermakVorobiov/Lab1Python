1. Створіть функцію, яка приймає назву файлу та назву стовпця як
параметр та повертає масив numpy зі значеннями стовпця.
2. Використовуючи функцію з попереднього завдання, створіть масив
numpy зі значень стовпця «APC» (Armored personnel carrier)..

```python
import csv
import numpy as np

def readTheColumn(file_path, column):
    with open(file_path, 'r', newline='') as f:
        csv_reader = csv.reader(f)
        header = next(csv_reader)
        column_index = header.index(column)
        column_values = np.array(list(csv_reader))[:, column_index]

    return column_values

APC_losses = readTheColumn('russia_losses_equipment.csv', 'APC')
print(APC_losses)
```

7. Створіть графік втрат БТР за останні 200 днів (починаючи з останньої
дати в наборі даних)

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

def readTheColumn(file_path, column):
    with open(file_path, 'r', newline='') as f:
        csv_reader = csv.reader(f)
        header = next(csv_reader)
        column_index = header.index(column)
        column_values = np.array(list(csv_reader))[:, column_index]

    return column_values

APC_losses = readTheColumn('russia_losses_equipment.csv', 'APC')
APC_losses = APC_losses.astype(int)
APC_daily_losses = np.concatenate(([APC_losses[0]], np.diff(APC_losses)))

plt.figure(figsize=(8, 16), dpi=100)
plt.plot(np.arange(426, 626), APC_daily_losses[np.arange(426, 626)], label='APC Losses', linestyle='--', marker='o', color='green')

plt.xlabel('Дата')
plt.ylabel('Кількисть одиниць')
plt.title('Графік втрат БТР-ів')


plt.legend()

plt.grid(True, linestyle='--', color='gray')  
plt.xticks(fontsize=10, rotation=90)
plt.yticks(fontsize=10)

plt.savefig('apclosses.png', dpi=100, bbox_inches='tight')

plt.show()
```
